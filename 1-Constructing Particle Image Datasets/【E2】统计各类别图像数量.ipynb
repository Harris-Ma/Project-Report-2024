### 图像分类数据集探索统计

### 各类别数据个数柱状图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### 导入数据集统计表格

```python
df = pd.read_csv('数据量统计.csv')
```

```python
df.shape
```

```python
df
```

### 图像数量柱状图可视化

```python
# 指定可视化的特征
feature = 'total'
# feature = 'trainset'
# feature = 'testset'
```

```python
df = df.sort_values(by=feature, ascending=False) #倒序
```

```python
df.head()
```

```python
plt.figure(figsize=(22, 7))

x = df['class']
y = df[feature]

plt.bar(x, y, facecolor='#1f77b4', edgecolor='k')

plt.xticks(rotation=90)
plt.tick_params(labelsize=15)
plt.xlabel('Class', fontsize=20)
plt.ylabel('Number', fontsize=20)

# plt.savefig('各类别图片数量.pdf', dpi=120, bbox_inches='tight')

plt.show()
```

```python
plt.figure(figsize=(22, 7))
x = df['class']
y1 = df['testset']
y2 = df['trainset']

width = 0.55 # 柱状图宽度

plt.xticks(rotation=90) # 横轴文字旋转

plt.bar(x, y1, width, label='Testset')
plt.bar(x, y2, width, label='Trainset', bottom=y1)


plt.xlabel('Class', fontsize=20)
plt.ylabel('Number', fontsize=20)
plt.tick_params(labelsize=13) # 设置坐标文字大小

plt.legend(fontsize=16) # 图例

# 保存为高清的 pdf 文件
plt.savefig('各类别图像数量2.pdf', dpi=120, bbox_inches='tight')

plt.show()
```
